```python
%%sql
SELECT distinct COUNT(*) AS NUMBER_OF_CRIMES 
FROM CHICAGO_CRIME_DATA;
```

```python
%%sql select COMMUNITY_AREA_NUMBER, COMMUNITY_AREA_NAME, PER_CAPITA_INCOME from CENSUS_DATA \
where PER_CAPITA_INCOME < 11000;
```

```python
%sql select CASE_NUMBER from CHICAGO_CRIME_DATA where DESCRIPTION LIKE '%MINOR';
```

```python
%sql select * from CHICAGO_CRIME_DATA where PRIMARY_TYPE = 'KIDNAPPING' and DESCRIPTION LIKE 'CHILD%';
```

```python
%%sql select distinct(PRIMARY_TYPE) from CHICAGO_CRIME_DATA
where LOCATION_DESCRIPTION LIKE 'SCHOOL%'
```

```python
%%sql     
select "Elementary, Middle, or High School",AVG(SAFETY_SCORE) as AVERAGE_SAFETY_SCORE
from CHICAGO_PUBLIC_SCHOOLS
group by "Elementary, Middle, or High School";
```

```python
%%sql select COMMUNITY_AREA_NAME, PERCENT_HOUSEHOLDS_BELOW_POVERTY from CENSUS_DATA \
Order by PERCENT_HOUSEHOLDS_BELOW_POVERTY DESC LIMIT 5;
```

```python
%%sql select COMMUNITY_AREA_NUMBER ,count(COMMUNITY_AREA_NUMBER) as FREQUENCY 
from CHICAGO_CRIME_DATA 
group by COMMUNITY_AREA_NUMBER 
order by count(COMMUNITY_AREA_NUMBER) desc 
limit 1;
```

```python
%%sql select COMMUNITY_AREA_NAME
from  CENSUS_DATA
where HARDSHIP_INDEX = (select MAX(HARDSHIP_INDEX) from CENSUS_DATA);
```

```python
%%sql
select COMMUNITY_AREA_NAME
from CENSUS_DATA 
where COMMUNITY_AREA_NUMBER = (    
    select CCD.COMMUNITY_AREA_NUMBER 
    from CHICAGO_CRIME_DATA AS CCD 
    group by CCD.COMMUNITY_AREA_NUMBER
    order by count(CCD.COMMUNITY_AREA_NUMBER) desc
    limit 1) limit 1;

```
